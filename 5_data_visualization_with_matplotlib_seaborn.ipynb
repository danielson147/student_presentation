##### Visualization Using Matplotlib

Matplotlib is a popular and widely used open-source data visualization library for the Python programming language. It was created by John D. Hunter in 2003 and has since become an essential tool for generating high-quality 2D and 3D plots and graphs.

```python
#import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Line plot
#let's say our x is fuel price and y is t-fare
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sample of Line Plot')
plt.show();


```

```python
# Scatter plot
plt.scatter(x, y, c='blue')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sample of Scatter Plot')
plt.show();

```

```python
# Bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 15, 10, 25, 30]
plt.bar(categories, values, color=['green'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show();
```

#### Customizing Plots

>

```python
# Customizing a plot
plt.scatter(x, y, c='red', marker='*', label='Data Points')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Nicely Customized Plot')
plt.legend()
plt.grid(True)
plt.show();
```

```python
# Load the 'iris' dataset from Seaborn
flower = sns.load_dataset('iris')

# Create a histogram of the 'sepal_length' column with 10 bins
# bins=10 specifies the number of bins or intervals 
# that the data will be divided into for the histogram. In this case, it's set to 10.
# bins=10 in this code specifies that the 'sepal_length' data will be divided into 10 equal-width 
# intervals for the histogram, allowing you to see how the data is distributed across these intervals.
sns.histplot(flower['sepal_length'], bins=10)

# Set the x-axis label
plt.xlabel('Sepal Length')

# Set the y-axis label
plt.ylabel('Frequency')

# Set the title of the plot
plt.title('Histogram')

# Display the plot on the screen
plt.show();
```

```python
sns.boxplot(x='species', y='petal_length', data=flower)
plt.xlabel('Species of Flower')
plt.ylabel('Flower Petal Length')
plt.grid(True)
plt.title('Box Plot by Species')
plt.show();
```

>The line inside the box represents the median (middle) value of the petal lengths for that species.
>
>The box itself represents the interquartile range (IQR), which spans from the first quartile (25th percentile) to the third quartile (75th percentile) of the petal lengths. It contains the central 50% of the data.
>
>The "whiskers" extend from the box, indicating the range of the data. They typically extend to a maximum of 1.5 times the IQR.
>
>Any individual data points that fall outside the whiskers are considered outliers and are represented as individual points.
