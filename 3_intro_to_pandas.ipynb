```python
# Selecting columns
age_column = df1['Age']

# Selecting rows
alice_data = df1[df1['Name'] == 'Alice']

# Slicing data
subset = df1.iloc[1:3, 0:2]  # Rows 1 to 2 and columns 0 to 1

# Display the selected data
print("Age Column:")
print(age_column)

print("\nAlice's Data:")
print(alice_data)

print("\nSubset of Data:")
print(subset)
```
